```python
import pandas as pd
import csv
```

```python
f = pd.read_csv('201903-citibike-tripdata.csv')
f.head()
```

```python
f['Duration_Minutes'] = f['tripduration'] / 60
f.head()
```

```python
dec = 2
f['Duration_Minutes'] = f['Duration_Minutes'].apply(lambda x: round(x, dec))
f.head()
```

```python
f.to_csv(r'clean_201903-citibike-tripdata.csv', index=None, header=True)
```
